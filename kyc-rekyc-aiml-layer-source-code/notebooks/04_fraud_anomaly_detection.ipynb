# Fraud Anomaly Detection
Isolation Forest baseline for abnormal KYC profiles.

```
import sys, pandas as pd
sys.path.append("..")
from scripts.generate_synthetic_kyc_data import build
df=build(50000)
FEATURES=["age","monthly_income","loan_amount","document_mismatch_count","sanctions_hit","pep_hit","adverse_media_hit","geo_distance_km","rekyc_overdue_days"]
```

```
from sklearn.ensemble import IsolationForest
model=IsolationForest(contamination=0.08, random_state=42).fit(df[FEATURES])
df["anomaly_score"]=-model.decision_function(df[FEATURES])
df.sort_values("anomaly_score", ascending=False).head(10)
```

```
import os, joblib
os.makedirs("../artifacts", exist_ok=True)
joblib.dump(model, "../artifacts/fraud_model.joblib")
```
